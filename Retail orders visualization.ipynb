```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set(style="ticks")
```

```python
path = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Online_Retail/Online_Retail.csv'
retail_purchase=pd.read_csv(path, encoding = 'latin1')
retail_purchase.head()
```

```python
filtered_retail_purchase = retail_purchase[retail_purchase['Country'] != 'UK']
grouped_retail_purchase = filtered_retail_purchase.groupby('Country')['Quantity'].sum().nlargest(10)
countries = grouped_retail_purchase.index
quantity_values = grouped_retail_purchase.values

# Create the histogram
plt.bar(countries, quantity_values)
plt.xlabel('Country')
plt.ylabel('Quantity')
plt.title('Top 10 Countries by Quantity Ordered (excluding UK)')
plt.xticks(rotation=45)
plt.show()
```

```python
countries = retail_purchase.groupby('Country').sum()


countries = countries.sort_values(by = 'Quantity',ascending = False)[1:11]

countries['Quantity'].plot(kind='bar')

plt.xlabel('Countries')
plt.ylabel('Quantity')
plt.title('10 Countries with most orders')

plt.show()
```

```python
grouped_retail_purchase = retail_purchase.groupby('Country')['Quantity'].sum().nlargest(10)
countries = grouped_retail_purchase.index
quantity_values = grouped_retail_purchase.values

plt.bar(countries, quantity_values)
plt.xlabel('Country')
plt.ylabel('Quantity')
plt.title('Top 10 Countries by Quantity Ordered')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 1: Line Plot
retail_purchase['InvoiceDate'] = pd.to_datetime(retail_purchase['InvoiceDate'])
plt.figure(figsize=(10, 6))
plt.plot(retail_purchase['InvoiceDate'], retail_purchase['Quantity'])
plt.xlabel('Invoice Date')
plt.ylabel('Quantity')
plt.title('Quantity Ordered Over Time')
plt.show()
```

```python
# Plot 2: Bar Plot
country_sales = retail_purchase.groupby('Country')['Quantity'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
plt.bar(country_sales.index, country_sales.values)
plt.xlabel('Country')
plt.ylabel('Quantity')
plt.title('Top 10 Countries by Quantity Ordered')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 3: Histogram
plt.figure(figsize=(10, 6))
plt.hist(retail_purchase['Quantity'], bins=30)
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.title('Distribution of Quantity')
plt.show()# Plot 3: Histogram
plt.figure(figsize=(10, 6))
plt.hist(retail_purchase['Quantity'], bins=30)
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.title('Distribution of Quantity')
plt.show()
```

```python
# Plot 4: Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(retail_purchase['UnitPrice'], retail_purchase['Quantity'])
plt.xlabel('Unit Price')
plt.ylabel('Quantity')
plt.title('Quantity vs. Unit Price')
plt.show()
```
